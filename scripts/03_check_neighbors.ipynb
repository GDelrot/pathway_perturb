```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
RESULTS = Path('/home/gdelrot/pathway_perturb/outputs/parallele_KNN')
```

```python
results = {}
for pert_type in ['raw_pert','pathway_pert']:
    results[pert_type] = {}
    for metadata in ['moa','cluster_id']:
        dfs = []
        for file in RESULTS.glob(f'*{pert_type}_{metadata}*'):
            if file.suffix.endswith('csv'):
                dfs.append(pd.read_csv(file, index_col=0))
        
        result_df = pd.concat(dfs).groupby(level=0).median()
        result_df.to_csv(f'test_{metadata}_{pert_type}.csv')
        results[pert_type][metadata] = result_df
```
